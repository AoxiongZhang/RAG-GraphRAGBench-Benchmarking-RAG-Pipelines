```python
from datasets import load_dataset

# Login using e.g. `huggingface-cli login` to access this dataset
dataset = load_dataset("galileo-ai/ragbench", "emanual", split="validation")

print(dataset.column_names)
```

```python
dataset[0]
```

```python
filtered = dataset.filter(lambda example: example["generation_model_name"] == "gpt-3.5-turbo-0125")
```

```python
print(filtered)
print(f"Original Number: {len(dataset)}, After filtering: {len(filtered)}")
```

```python
import pandas as pd

rows = []

for item in filtered:

    user_input = item["question"]
    reference  = item["response"]
    rows.append({
        "user_input": user_input,
        "reference":  reference,
        "context":    "",
        "answer":     ""
    })

df = pd.DataFrame(rows)
print(df.head())
```

```python
df.to_csv("evaluate_dataset/test_dataset_ragbench.csv")
```
